```python
shape = ['원형', '장방형', '타원형', '사각형', '기타', '팔각형', '오각형', '삼각형', '마름모형','육각형', '반원형']
```

```python
df = pd.read_csv('./class_no.csv')
```

```python
df1 = df[['품목일련번호','제형코드명','의약품제형','색상앞','색상뒤']]
```

```python
df1 = df1.drop_duplicates('품목일련번호')
```

```python
df1.set_index('품목일련번호',inplace=True)
```

```python
src_path = './ori_resize'
dst_path = './drug2'
```

```python
drugfiles = os.listdir(src_path)
if 'pill' not in os.listdir(dst_path):
    os.mkdir(dst_path+'/pill')
    os.mkdir(dst_path+'/capsule')
    os.mkdir(dst_path+'/undef')
```

```python
for drugfile in drugfiles:
    label = int(drugfile.split('-')[0])
    if df1.loc[label,'제형코드명'] in pill:
        shutil.copyfile(f'{src_path}/{drugfile}', f'{dst_path}/pill/{drugfile}')
    elif df1.loc[label,'제형코드명'] in capsule:
        shutil.copyfile(f'{src_path}/{drugfile}', f'{dst_path}/capsule/{drugfile}')
    elif df1.loc[label,'제형코드명'] in undef:
        shutil.copyfile(f'{src_path}/{drugfile}', f'{dst_path}/undef/{drugfile}')
```

```python
src_path = './drug/undef'
dst_path = './drug2/undef'
drugfiles = os.listdir(src_path)
if 'one' not in os.listdir(dst_path):
#     os.mkdir('./drug3')
    os.mkdir(dst_path+'/one')
    os.mkdir(dst_path+'/jangbang')
    os.mkdir(dst_path+'/taone')
    os.mkdir(dst_path+'/sagack')
    os.mkdir(dst_path+'/gita')
    os.mkdir(dst_path+'/pargack')
    os.mkdir(dst_path+'/fivegack')
    os.mkdir(dst_path+'/trigack')
    os.mkdir(dst_path+'/marumo')
    os.mkdir(dst_path+'/sixgack')
    os.mkdir(dst_path+'/banone')
```
